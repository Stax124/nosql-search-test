```python
index.delete_all_documents()
```

```python
# Add documents to the index
res = requests.get("https://raw.githubusercontent.com/voidful/gpu-info-api/gpu-data/gpu.json")
data = res.json()
data
```

```python
data_arr = []
for key, value in data.items():
    coverted = convert_keys(value)
    data_arr.append(coverted)
    
data_arr
```

```python
for i, value in enumerate(data_arr):
    data_arr[i]["id"] = i
```

```python
data_arr
```

```python
task = index.add_documents(data_arr, primary_key="id")
client.wait_for_task(task.task_uid)
```

```python
# Index properties on documents
task = index.update_filterable_attributes([
    "id",
    "model",
    "launch_time",
    "vendor",
    "gpu_type",
])
client.wait_for_task(task.task_uid)
```
